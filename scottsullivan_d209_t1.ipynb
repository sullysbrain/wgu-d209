## Part II: Method Justification
### B1. Classification Analysis
The author chose to create a Naive Bayes classification model for this task. The Bayes classifier works by taking the probabilites of each scenerio and combining them into a model to predict the probability of a specific event. The Bayes model was also chosen because at larger data sizes, it is faster than K-Nearest Neighbor (source: Glen).  In this case, the event is the dependent attribute being true, or the customer churning.

### B2. Assumptions
The Naive Bayes classifyer does have an assumption built into it. It assumes that the effects of a specific feature is fully independent from the other features. For example, if the independent variables include age and income (such as our model), the classifer assumes that age and income are not interdependent (ie. one's income increases in relationship to their age). In real life, there are many interdependent factors, but for this example, we will assume that all the attributes are fully independent.

## Part III: Data Prep
### C1. Preprocessing Goal
One of the preprocessing goals required for analyzing the 'Churn' dataset will be to transform the Yes/No and True/False categorical attributes to a numerical 1/0, respectively. In the same way, multi-categorical attributes (such as InternetService) will need to have the same transformation run, except that since they are not binary in nature, the author will use One Hot Encoding to do so, which will prevent multicollinearity. 

But prior to these transformations, the author will do the standard data cleaning procedures which include dropping the non-useful attributes (like customer UUID, CaseNumber, etc) and then checking for duplicates, null values, and outliers.

#### CSV Import

```python
import numpy as np
import pandas as pd

# Read raw CSV File
nulls = ['nan', 'NaN', '']
f_path = '_data/churn_clean.csv'
df = pd.read_csv(f_path, keep_default_na=False, na_values=nulls)

# Copy DF just in case we need the raw data
df_rawCSV = df
df.head(3)
```

```python
# Dropping less meaningful attributes
cols_to_drop = ['CaseOrder', 'Customer_id', 'Interaction', 'Population', 'UID', 
                   'City', 'State', 'County', 'Zip', 'Lat', 'Lng', 'Area', 'Marital',  
                   'Job', 'TimeZone']
for co in df.columns:
    if co in cols_to_drop:
        df.drop(co, axis=1, inplace=True)

print(f"\nAttributes Left Over:\n{df.dtypes}")
```

#### Numerical Encoding of Binary Attributes

```python
binary_attr = ['Techie', 'Churn', 'Phone','Tablet','Port_modem', 
               'Multiple', 'DeviceProtection','OnlineSecurity',
               'OnlineBackup','StreamingMovies', 'PaperlessBilling',
               'StreamingTV','TechSupport', ]

for attString in binary_attr:
    df[attString] = df[attString].replace({"Yes": 1, "No": 0, "True": 1, "False": 0})
    df[attString] = df[attString].astype(np.int64)
    print(f"Changed: {attString}\t")
    
df.head(5)
```

#### Check for Duplicates

```python
dup_entries = len(df) - len(df.duplicated())
print(f"Duplicate Entries: {dup_entries} duplicates, of {len(df)} total.\n")
```

#### Check for Null Values

```python
df.isnull().sum()
```

#### Check for Outliers

```python
# Inspect using histograms for outliers (skip binary attributes)
%matplotlib inline
import matplotlib.pyplot as plt
from plotnine import ggplot, aes, geom_histogram
import seaborn as sns

df_attr_to_plot = pd.DataFrame()
for attribute in df.columns:
    if attribute not in binary_attr:
        df_attr_to_plot[attribute] = df[attribute]
        
fig, axs = plt.subplots(nrows=8, ncols=3, figsize=(18,16))
for ax, attr in zip(axs.flatten(), df_attr_to_plot):
    sns.histplot(data=df, x=attr, bins=25, color='#3bb3a6', ax=ax)
    ax.set_title(attr)
    ax.set_xlabel(attr)
    ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

#### Rename Attributes to More Logical Ones

```python
to_rename = {'Item1':'TimelyResponse', 'Item2':'TimelyFixes','Item3':'TimelyReplacements',
                  'Item4':'Reliability','Item5':'Options','Item7':'Courteous',
                  'Item6':'Respectful','Item8':'ActiveListening'}

df = df.rename(columns=to_rename)
print('Renamed')
```

#### One Hot Encoding

```python
# One Hot Encoding
one_hot_list = ['InternetService', 'Contract', 'PaymentMethod', 'Gender']

for item in one_hot_list:
    if item in df.columns:
        # Added astype to convert from boolean to numerical
        y = pd.get_dummies(df[item], prefix=item, drop_first=True).astype(np.int64)
        df.drop(item, axis=1, inplace=True)
        df = pd.concat([df, y], axis="columns")

print(df.head(5))
```

## Part IV: Analysis
### D1. Split Data (Training and Testing sets)
The data will be split into both a training and testing set.

```python
# Create a dataframe of ONLY independent variables (ie. with 'Churn' dropped)
ind_variable_df = pd.DataFrame()
ind_variable_df = df.drop('Churn', axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, f1_score

X = ind_variable_df
y = df['Churn']

# Build test and train sets
X_tr, X_te, y_tr, y_te = train_test_split(
    X, y, test_size=0.3, random_state=3
)
```

```python
# Export Training and Testing Datasets to CSV
file_name_Xtrain = '_data/d209_t1_Xtrain_data.csv'
file_name_ytrain = '_data/d209_t1_ytrain_data.csv'
file_name_Xtest = '_data/d209_t1_Xtest_data.csv'
file_name_ytest = '_data/d209_t1_ytest_data.csv'

X_tr.to_csv(file_name_Xtrain, index=False)
y_tr.to_csv(file_name_ytrain, index=False)
X_te.to_csv(file_name_Xtest, index=False)
y_te.to_csv(file_name_ytest, index=False)

print("Exported training and testing files.")
```

### D2. Analysis Description
The author will create a Naive Bayes model using Scikit-learn. This was accomplished using Scikit-learn, and the author chose a Gaussian Naive Bayes classification (source: Awan). In order to accomplish this, the author took the training set (comprised of X_tr, y_tr) and created in the previous section and used scikit-learn to fit the model, training it. 

As an exploratory measure, the author then randomly selected a number of entries from the test set and used the model to predict a value (whether or not the customer churned), and compared the generated prediction with the actual value. The code loops through each of these test values to allow further exploration on the data and model accuracy (though a more precice accuracy analysis will be conducted later).

### D3. Perform Classification Analysis

Below, the author created the Naive Bayes model, trained it, and also ran several random predictions against the test values to explore at a high level the accuracy of the model (note: this is NOT the same as running an actual accuracy test, which the author does perform in section E).

```python
# Build Naive Bayes model
nb_model = GaussianNB()

# Model training
nb_model.fit(X_tr.values, y_tr)

# check against random test set
number_of_tests = 25
import numpy as np
for i in range(0,number_of_tests):
    rand_ind = np.random.randint(0,len(X_te))
    pred = nb_model.predict([X_te.iloc[rand_ind]])
    answer = "Correct" if pred[0] == y_te.iloc[rand_ind] else "Wrong"
    print(f"Actual value at {rand_ind}:\t{y_te.iloc[rand_ind]}\tPredected: {pred[0]}\t{answer}")
```

## Part V: Data Summary
### E1. Accuracy and the area under the curve (AUC)
The ROC curve (and the Area Under the Curve) measures how good the test is overall. A larger the area, the more useful the test. The AUC indicates how much the model can tell the difference between classes. In a perfect world, our model should have an AUC as close to 1 as possible. When we see an AUC near 0, we can tell the model is of poor quality (source: Bowne-Anderson)

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Predicted probabilities
y_pred_prob = nb_model.predict_proba(X_te.values)[:,1]

# ROC curve calculation
fpr, tpr, _ = roc_curve(y_te, y_pred_prob)

# Plotting the ROC curve
plt.plot(fpr, tpr, label='ROC curve (AUC = %0.2f)' % roc_auc_score(y_te, y_pred_prob))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.01])
plt.xlabel('False Positive')
plt.ylabel('True Positive')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```

```python
# Calculate accuracy score
y_pr = nb_model.predict(X_te.values)
accuracy = accuracy_score(y_pr, y_te)
f1 = f1_score(y_pr, y_te, average="weighted")

print("Accuracy:", accuracy)
print("F1 Score:", f1)
```

As we can see, the accuracy is decent. Though in the author's opinion, if time were available before the hypothetical stakeholder meeting, he would prefer to dial in the model more, refitting it to have a higher degree of accuracy.
